````python
def find_descendent_nodes(node, node_dependencies, skip_nodes):
    node_set = set()
    queue = [node]
    while queue:
        parent_node = queue.pop()
        if parent_node in skip_nodes: continue
        for child in node_dependencies[parent_node]:
            if child not in node_set:
                node_set.add(child)
            queue.append(child)
            print(f"Queue: {queue}, Child: {child}, Parent_node:{node_dependencies[parent_node]}")

    return node_set
````
